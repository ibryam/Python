```python
from random import randint

EASY_LEVEL_TURNS = 10
HARD_LEVEL_TURNS = 5



def guess_result(guess, random_number, attempts):
    if guess > random_number:
        print("Too high.▼")
        print("Try again!")
        return attempts - 1
    elif guess < random_number:
        print("Too low.▲")
        print("Try again!")
        return attempts - 1
    else:
        print(f"You got guessed it. The number is ☑☑☑ {random_number} ☑☑☑ !")
        
        
def set_difficulty():
    level = str(input("Choose a difficulty. Type 'easy' or 'hard': "))
    if level == 'easy':
        return EASY_LEVEL_TURNS
    else:
        return HARD_LEVEL_TURNS




def game():
    print("Welcome to the Number Guess Game!")
    print("I am thinking of a numner netweem 1 amd 100.")
    random_number = (randint(1, 100))
#     print(f"shh the number is: {random_number}")
    attempts = set_difficulty()
    

    guess = 0
    while guess != random_number:
        print(f"You have {attempts} attempts remaining to guess the number.")
        guess = int(input("Make a guess: "))

        attempts = guess_result(guess, random_number, attempts)
        if attempts == 0:
            print("You've run out of guesses. You lose!")
            return

game()
```
